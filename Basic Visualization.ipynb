### Saving figures to file

```python
#Let's create our first graph

x = np.linspace(0, 10, 100)

fig = plt.figure()    #To create a plot figure

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

#To show our plot, we use show()
plt.show() 
```

```python
#savefig() function saves the plots in our working directory

fig.savefig('myfirstviz.png')
```

### Multiple Plots

```python
#To create two plots

# create a plot figure
plt.figure()

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
#To create three plots

# create a plot figure
plt.figure()

# create the first of three panels and set current axis

plt.subplot(3, 1, 1)    #(rows, columns, panel number)
plt.plot(x, np.sin(x))  #To plot the data

# create the second panel and set current axis
plt.subplot(3, 1, 2)
plt.plot(x, np.cos(x));

# create the third panel and set current axis
plt.subplot(3, 1, 3)
plt.plot(x, np.sin(x));
```

For all plots, we first start by creating figure and axis

### Line Color

```python
#Line colors (, color = _)

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1. 0 - Dark Grey
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
```

### Line Types and Colors

```python
#Line Style can be adjusted by (, linestyle = _)

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 2, linestyle='dashed')
plt.plot(x, x + 4, linestyle='dashdot')
plt.plot(x, x + 6, linestyle='dotted');
```

```python
# For short, you can use the following codes:

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 6, linestyle='--') # dashed
plt.plot(x, x + 8, linestyle='-.') # dashdot
plt.plot(x, x + 10, linestyle=':');  # dotted
```

```python
#To combine line style and color

plt.plot(x, x + 0, '-g')   # solid green
plt.plot(x, x + 1, '--r')  # dashed red
plt.plot(x, x + 2, '-.b')  # dashdot blue
plt.plot(x, x + 3, ':y');  # dotted yellow
```

### Axes Limits in a Plot

```python
#To change the axes of our plots we use xlim and ylim

plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
#To combine x and y axes

plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])    #[xmin, xmax, ymin, ymax]
```

```python
#To automatically set axes and tighten the bounds around the current plot

plt.plot(x, np.sin(x))
plt.axis('tight');
```

### Labeling Plots

```python
#Title - plt.title(' ')

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
```

```python
#To label the axes - plt.xlabel() and plt.ylabel()

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
# Plot Legend

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.legend();
```

```python
#Customizing plot legend

plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')

plt.legend(loc = 'lower right', frameon = False);

#loc is used for the location of legend
#frameon is for the outer frame (True or False)
```

### Scatter Plots

```python
#plt.plot is used for scatterplots

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')      #o is the type of symbol
```

```python
# Different types of symbols

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
```

```python
#To combine different types of line and symbol

plt.plot(x, y, '-o');
```

```python
#plt.scatter is another way to plot

plt.scatter(x, y, marker = 'o');
```

## Histograms

- A histogram can be a great first step to understand the distribution of the data

```python
#plt.hist is used to plot a histogram

df = np.random.randn(1000)
plt.hist(df);
```

```python
#Creating bins

plt.hist(df, bins=30, normed=True, histtype='stepfilled', color = 'steelblue', alpha = 0.5)

#alpha is used for the strength of the color and it ranges from 0 to 1
#histtype - bar, barstacked, step, stepfilled
```

- 'bar' is a traditional bar-type histogram. If multiple data are given the bars are arranged side by side.
- 'barstacked' is a bar-type histogram where multiple data are stacked on top of each other.
- 'step' generates a lineplot that is by default unfilled.
- 'stepfilled' generates a lineplot that is by default filled.

# Seaborn

Seaborn provides an API on top of Matplotlib that offers multiple choices for plot style and color defaults, defines simple high-level functions for common statistical plot types

The main idea of Seaborn is that it provides high-level commands to create a variety of plot types useful for statistical data exploration, and even some statistical model fitting.

```python
import seaborn as sns
```

```python
#Matplotlib Histogram

data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], normed=True, alpha=0.5)
```

- For the same data, we can have a smooth estimate of the distribution using a kernel density estimate (kde)

```python
#sns.kdeplot

for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
#Histograms and KDE can be combined using distribution plots (distplot)

sns.distplot(data['x'])
sns.distplot(data['y']);
```
