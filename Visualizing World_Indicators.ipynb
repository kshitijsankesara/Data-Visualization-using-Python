# Data Visualization using Matplotlib and Seaborn

```python
#Importing matplotlib and seaborn

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
#Barplot showing average ease of business for each region

plt.figure(figsize = (12,8))                               #Setting size of the figure (12 - Width, 8 - Height)
sns.barplot("Region", "Ease of Business Index", data=df) 
plt.title('Ease of Business for Regions')                   #Title
plt.xlabel('Region', fontsize=14)                              #Label for X axis
plt.ylabel('Ease of Business Index', fontsize=12)              #Label for Y axis
plt.show()                                                   #To print the plot
```

```python
#Pie Chart showing count of Regions

plt.figure(figsize = (12,8))
df.Region.value_counts().plot.pie()
plt.axis('equal')                        #Sets and adjust plots with equal size ratio
plt.show()
```

```python
#Line Plot - To Plot Regions and Sum of their Population

plt.figure(figsize = (12,8))
df.groupby(['Region']).sum()['Population Total (Millions)'].plot(color = 'red')
plt.ylabel('Population (Millions)', fontsize=14)
plt.title('Sum of Population of each Region')
plt.show()
```

```python
#Scatter Plot

df.plot.scatter('Health Exp/Capita ($)', 'GDP( Billions)', alpha = 0.5, color = 'green')
plt.show()

#Alpha is between 0 and 1. O being Transparent and 1 being Opaque
```

```python
#Heatmap for checking the correlation between variables using seaborn
```

```python
plt.figure(figsize = (12,8))
corrplot = sns.heatmap(df.corr())
corrplot.set_xticklabels(corrplot.get_xticklabels(), rotation = 45, horizontalalignment = 'right') #For X axis labels
plt.show()
```

```python
#Boxplot for showing Region and Infant Mortality Rate

plt.figure(figsize = (12,8))
sns.boxplot(x="Region", y="Infant Mortality Rate", data=df, palette="Set2")    #Palette are used for choosing colors
plt.show()
```

```python
#Boxplot for showing Region and Days for starting Business

plt.figure(figsize = (12,8))
sns.boxplot(x="Region", y="Days to Start Business", data=df, palette="Set1")    
plt.show()
```

```python
#ViolinPlot to plot region with hours required for tax preparation
plt.figure(figsize = (12,8))
sns.violinplot(x="Region", y="Hours to Prepare Tax", data=df, palette="Set2")
plt.show()
```

```python
#Bar plot showing Inbound Tourism for each Region

plt.figure(figsize = (12,8))
plt.bar(x = "Region", height = "Tourism Inbound($Billion)", data = df, color = 'gray')
plt.show()
```

```python
#Bar plot showing Outbound Tourism for each Region

plt.figure(figsize = (12,8))
plt.bar(x = "Region", height = "Tourism Outbound($Billion)", data = df, color = 'red')
plt.show()
```
