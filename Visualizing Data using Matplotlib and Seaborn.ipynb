# Data Visualization using Matplotlib and Seaborn

```python
#Importing library

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Reading csv file

data = pd.read_csv('gapminder.tsv', delimiter='\t')
data.head()
```

## Different Line Styles

```python
#Plotting different lines
#Line Style can be adjusted by (, linestyle = _)

x = np.linspace(0, 10, 100)                    #Generating random numbers
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 2, linestyle='dashed')
plt.plot(x, x + 4, linestyle='dashdot')
plt.plot(x, x + 6, linestyle='dotted');
```

```python
# For short, you can use the following codes:

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 6, linestyle='--') # dashed
plt.plot(x, x + 8, linestyle='-.') # dashdot
plt.plot(x, x + 10, linestyle=':');  # dotted
```

## Different Line Colors

```python
#Line colors (, color = _)

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1. 0 - Dark Grey
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
```

### To combine line style and color

```python
#To combine line style and color

plt.plot(x, x + 0, '-g')   # solid green
plt.plot(x, x + 1, '--r')  # dashed red
plt.plot(x, x + 2, '-.b')  # dashdot blue
plt.plot(x, x + 3, ':y');  # dotted yellow
```

## 1. Line Plot

```python
#Now let's plot two columns of data frame, here we are creating a line plot of gdpPercap and lifeExp

plt.plot(data.gdpPercap, data.lifeExp)
plt.show()
```

## 2. Scatter Plot

```python
# Scatter Plot showing gdpPercap and lifeExp

plt.scatter(data.gdpPercap, data.lifeExp, color = 'blue')
plt.show()
```

```python
# Scatter Plot, adding labels and titles

plt.scatter(data.gdpPercap, data.lifeExp, color = 'blue')
plt.xlabel('gdp per Capita in USD')                        #X Label
plt.ylabel(' Life Expectancy')                             #Y Label
plt.title('World Life Expectancy vs GDP')                  #Title
plt.show()
```

## 3. Histogram

```python
#Histogram showing distribution of LifeExp

plt.hist(data.lifeExp, 25)                                #25 is the number of bins                           
plt.xlabel('Life Expectancy')                             #X Label
plt.ylabel('Distribution')                                #Y Label
plt.title('Distribution of World Life Expectancy')        #Title
plt.show()
```

## 4. Barplot

```python
#Barplot showing continent and population

plt.figure(figsize = (12,8))                               #Setting size of the figure (12 - Width, 8 - Height)
sns.barplot("continent", "pop", data=data)
plt.show() 
```

```python
#Horizontal Barplot showing continent and population

plt.figure(figsize = (12,8))                               
sns.barplot("pop", "continent", data=data)        #We are changing the X and Y axis to plot horizontal barchart
plt.show() 
```

```python
#Barplot showing continent and population with different years

plt.figure(figsize = (12,8))                               #Setting size of the figure (12 - Width, 8 - Height)
sns.barplot("continent", "pop", data=data, hue = 'year')   #We use hue = '' for encoding different colors in x axis
plt.show() 
```

## 5. Pie Chart

```python
#Pie Chart showing count of Continent

plt.figure(figsize = (12,8))
data.continent.value_counts().plot.pie()     #Using value_counts() for frequency of continent
plt.axis('equal')                            #Sets and adjust plots with equal size ratio
plt.show()
```

## 6. Line Plot

```python
#Line Plot - To Plot Continent and Sum of their Population using groupby function

plt.figure(figsize = (12,8))
data.groupby(['continent']).sum()['pop'].plot(color = 'red')    #Grouping by Continent and plotting sum of population
plt.title('Sum of Population of each continent')                #Title
plt.show()
```

## 7. Heat Map for Correlation Plot

```python
plt.figure(figsize = (12,8))
corrplot = sns.heatmap(data.corr())         #Making a heatmap for correlation between variables
plt.show()
```

## 8. Boxplot

```python
#Boxplot for showing continent and lifeExp
#Boxplot is a graph that presents distribution of the variable
#It is used to show the distribution and central value

plt.figure(figsize = (12,8))
sns.boxplot(x="continent", y="lifeExp", data=data, palette="Set2")    #Palette are used for choosing colors
plt.show()
```

## 9. ViolinPlot

```python
#ViolinPlot to plot Continent with life expactancy
#It is a method of plotting numerical data and it is similar to a box plot
#It shows a kernel density plot on each side

plt.figure(figsize = (12,8))
sns.violinplot(x="continent", y="lifeExp", data=data, palette="Set2")
plt.show()
```
